```python
import os
import torchaudio
from speechbrain.inference.TTS import Tacotron2
from speechbrain.inference.vocoders import HIFIGAN
from IPython.display import Audio, display

# =======================
# Initialize TTS (Tacotron2) and Vocoder (HiFi-GAN)
# =======================
tacotron2 = Tacotron2.from_hparams(source="speechbrain/tts-tacotron2-ljspeech", savedir="tmpdir_tts")
hifi_gan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-ljspeech", savedir="tmpdir_vocoder")

def chat():
    print("Welcome to the TTS Chat! Type 'exit' to quit.\n")
    while True:
        user_text = input("You: ")
        if user_text.lower() == "exit":
            print("Goodbye!")
            break

        # Generate mel-spectrogram from the user text
        mel_output, mel_length, alignment = tacotron2.encode_text(user_text)
        # Generate waveform audio from the mel-spectrogram
        waveforms = hifi_gan.decode_batch(mel_output)

        # Save the output audio file
        output_filename = "output.wav"
        torchaudio.save(output_filename, waveforms.squeeze(1), 22050)
        print(f"Audio saved to {output_filename}\n")

        # Play the generated audio in Colab
        display(Audio(output_filename, rate=22050))

if __name__ == "__main__":
    chat()
```

```python
import torch
import torchaudio
from scipy.io.wavfile import write
from IPython.display import Audio
```

```python
#Load the Tacotron2 model pre-trained on LJ Speech dataset and prepare it for inference:
tacotron2 = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tacotron2', model_math='fp16')
tacotron2 = tacotron2.to('cuda')
tacotron2.eval()
```

```python
# Load pretrained WaveGlow model
waveglow = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_waveglow', model_math='fp16')
waveglow = waveglow.remove_weightnorm(waveglow)
waveglow = waveglow.to('cuda')
waveglow.eval()
```

```python
# Now, let's make the model say:
text = "Hello I am Nishchay"
```

```python
# Format the input using utility methods
utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tts_utils')
sequences, lengths = utils.prepare_input_sequence([text])
```

```python
# Run the chained models:
with torch.no_grad():
    mel, _, _ = tacotron2.infer(sequences, lengths)
    audio = waveglow.infer(mel)
audio_numpy = audio[0].data.cpu().numpy()
rate = 22050
```

```python
write("audio.wav", rate, audio_numpy)

Audio(audio_numpy, rate=rate)
```
